```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
keras.__version__, Sequential.__version__
```

```python
x_train = pd.read_csv('/content/final_train.csv')
y_train = pd.read_csv('/content/final_y_data.csv')
x_test = pd.read_csv('/content/final_predict.csv')
y_test = pd.read_csv('/content/final_y_predict.csv')
```

```python
len(x_train)
```

```python
len(y_train)
```

```python
model = Sequential()

model.add(Dense(64, input_dim=39, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(64, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(64, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(64, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(64, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(64, activation='relu'))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()
```

```python
hist = model.fit(x_train, y_train,
          epochs=100,
          batch_size=32,
          validation_split=0.25,
          # validation_data=0.1
          )
```

```python

score = model.evaluate(x_test, y_test, batch_size=200)
score
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, loss_ax = plt.subplots()
acc_ax = loss_ax.twinx()

loss_ax.plot(hist.history['loss'], 'y', label = 'train loss')
# loss_ax.plot(hist.history['val_loss'], 'r', label = 'val loss')

acc_ax.plot(hist.history['accuracy'], 'b', label = 'train accuracy')
# acc_ax.plot(hist.history['val_accuracy'], 'g', label = 'valid accuracy')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuracy')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```
