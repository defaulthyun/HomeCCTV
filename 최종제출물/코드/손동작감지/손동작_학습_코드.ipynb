```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
```

```python
# 테스트 데이터(기본_손동작_데이터.csv)를 코랩 파일에 넣고 테스트 해주세요
data = pd.read_csv('/content/기본_손동작_데이터.csv')

# 데이터를 학습시키기 편하게 전처리
y = data['label'].values
X = data.drop( ['label'], axis=1 ).values

df_labels = data['label'].nunique()

# 정답 레이블을 라벨 인코딩
encoder = LabelEncoder()
y = encoder.fit_transform(y)
y = pd.get_dummies(y).values
```

```python
model = Sequential()
model.add(Dense(30, input_dim=15, activation='relu'))
model.add(Dense(60, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(df_labels, activation='softmax'))

model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(X, y, batch_size=16, epochs=100)

model.save('model/손동작_감지_모델.h5')
```
